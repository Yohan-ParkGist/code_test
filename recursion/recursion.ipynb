```python
def factorial(x):
    if x <= 1:
        return 1
    return x*factorial(x-1)
    
x = int(input())
print(factorial(x))
```

```python
def fibonacci(x):
    if x<=1:
        return x
    return fibonacci(x-1)+fibonacci(x-2)
x = int(input())
print(fibonacci(x))
```

```python
import sys

def recursion(s, l, r):
    if l >= r: return 1, l+1
    elif s[l] != s[r]: return 0, l+1
    else: return recursion(s, l+1, r-1)

def isPalindrome(s):
    return recursion(s, 0, len(s)-1)

n = int(sys.stdin.readline())

for _ in range(n):
    word = sys.stdin.readline().rstrip()
    a, b = isPalindrome(word)
    print(a, b)
```
