```python
import sys

def power(a, b):
    if b == 1:
        return a%p
    
    if b%2 == 0:
        return ((power(a, b//2) % p)**2)%p
    else:
        return a*((power(a, (b-1)//2) % p)**2)%p

n, k = map(int, sys.stdin.readline().split())
p = 1000000007
factorial = [1]*(n+1)
for i in range(2, n+1):
    factorial[i] = (factorial[i-1]*i)%p

denominator = (factorial[k]*factorial[n-k])%p

print((factorial[n]*power(denominator, p-2))%p)
```
