```python
import sys

def minimum_upper_bound(lis, start, end, key):
    pos = end
    while start <= end:
        mid = (start + end) // 2
        if lis[mid] >= key:
            pos = mid
            end = mid - 1
        else:
            start = mid + 1
    return pos

n = int(sys.stdin.readline())
num_list = list(map(int, sys.stdin.readline().split()))

lis = [num_list[0]]
for i in range(1, n):
    if num_list[i] > lis[-1]:
        lis.append(num_list[i])
    else:
        key = num_list[i]
        replaceindex = minimum_upper_bound(lis, 0, len(lis)-1, key)
        lis[replaceindex] = key

print(len(lis))
```
