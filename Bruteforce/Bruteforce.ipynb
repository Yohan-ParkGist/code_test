```python
while 1:
    try:
        n, m = map(int, input().split())
        arr = list(map(int, input().split()))
        max = 0
        for i in range(n-2):
            for j in range(i+1, n-1):
                for k in range(j+1, n):
                    sum = arr[i]+arr[j]+arr[k]
                    if sum > max and sum <= m:
                        max = sum
        print(max)
        
    except Exception:
        break
```

```python
while 1:
    try:
        n = int(input())
        cnt = 0
        for i in range(n):
            if i+sum(list(map(int, str(i)))) == n:
                cnt = 1
                print(i)
                break
        if cnt ==0 :
            print(0)
            
    except Exception:
        break
```

```python
while 1:
    try:
        arr = list(map(int, input().split()))
        for x in range(-999, 1000):
            for y in range(-999, 1000):
                if arr[0]*x+arr[1]*y == arr[2] and arr[3]*x+arr[4]*y == arr[5]:
                    print(x, y)
            
    except Exception:
        break
```

```python
while 1:
    try:
        n, m = map(int, input().split())
        arr = []
        for _ in range(n):
            arr.append(input())
        # print(arr)
        minumum_change = 64 
        for i in range(n-7):
            for j in range(m-7):
                count = 0
                for k in range(i, i+8):
                    for l in range(j, j+8):
                        if (k+l)%2==0:
                            if arr[k][l]=="W":
                                count+=1
                        else:
                            if arr[k][l]=="B":
                                count+=1
                count = min(count, 64-count)
                minumum_change = min(minumum_change, count)
                
        print(minumum_change)
    except Exception:
        break
```

```python
while 1:
    try:
        n = int(input())
        bags = [3, 5]
        arr = [5000]*(n+1)
        for i in range(1, n+1):
            if i in bags:
                arr[i]=1
            else:
                for bag in bags:
                    if i-bag > 0:
                        arr[i]=min(arr[i], arr[i-bag]+arr[bag])
        if arr[n]==5000:
            print(-1)
        else:
            print(arr[n])

    except Exception:
        break
```
