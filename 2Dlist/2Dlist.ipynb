```python
while 1:
    try:
        arr1=[]
        arr2=[]
        A, B = map(int, input().split())
        for i in range(A):
            arr1.append(list(map(int, input().split())))

        for i in range(A):
            arr2.append(list(map(int, input().split())))

        for i in range(A):
            for j in range(B):
                print(arr1[i][j] + arr2[i][j], end=" ")
            print()
        
    except Exception:
        break
```

```python
while 1:
    try:
        arr=[]
        for _ in range(9):
            arr.append(list(map(int, input().split())))
        
        max = 0
        max_i = 0
        max_j = 0
        for i in range(9):
            for j in range(9):
                if max <= arr[i][j]:
                    max = arr[i][j]
                    max_i = i+1
                    max_j = j+1
        print(max)
        print(max_i, max_j)
                
    except Exception:
        break
```

```python
while 1:
    try:
        arr=[]
        for _ in range(5):
            arr.append(input())
        
        max = 0
        max_i = 0
        max_j = 0
        for i in range(9):
            for j in range(9):
                if max <= arr[i][j]:
                    max = arr[i][j]
                    max_i = i+1
                    max_j = j+1
        print(max)
        print(max_i, max_j)
                
    except Exception:
        break
```

```python
while 1:
    try:
        arr = [['']*15 for _ in range(15)]
        
        for i in range(5):
            word = input()
            for j in range(len(word)):
                arr[i][j] = word[j]
        
        for j in range(15):
            for i in range(5):
                if arr[i][j]!='':
                    print(arr[i][j], end='')
        
    except Exception:
        break
```

```python
while 1:
    try:
        arr = [[0]*100 for _ in range(100)]
        n = int(input())
        count = 0
        
        for _ in range(n):
            x, y = map(int, input().split())
            for i in range(x, x+10):
                for j in range(y, y+10):
                    if arr[i][j] == 0:
                        arr[i][j] = 1
                        count+=1
        print(count)
        
    except Exception:
        break
```
