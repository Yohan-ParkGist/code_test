```python
import sys

n = int(sys.stdin.readline())
num = sorted(list(map(int, sys.stdin.readline().split())))
x = int(sys.stdin.readline())

i = 0
j = n-1
count = 0
while i<j:
    
    temp = num[i] + num[j]
    if temp == x:
        count+=1
        i+=1
    elif temp < x:
        i+=1
    elif temp > x:
        j-=1
        
print(count)
```

```python
import sys

n, s = map(int, sys.stdin.readline().split())
num = list(map(int, sys.stdin.readline().split()))

i = 0
j = 0
short = n+1
current = num[0]
while 1:
    if current >= s:
        short = min(short, (j-i+1))
        current-=num[i]
        i+=1
    elif j == n-1:
        break
    else:
        j+=1
        current+=num[j]
        
if short == n+1:
    print(0)
else:
    print(short)
```

```python
import sys

n = int(sys.stdin.readline())

if n == 1:
    print(0)
elif n == 2 or n == 3:
    print(1)
else:
    is_prime = [True]*(n+1)
    is_prime[0] = is_prime[1] = False

    lim = int(n**0.5)
    for i in range(2, lim+1):
        if is_prime[i]:
            is_prime[i*i:n+1:i] = [False]*len(range(i*i, n+1, i))
    
    num = [i for i in range(2, n+1) if is_prime[i]]
    
    left = 0
    right = 0
    count = 0
    short = n+1
    length = len(num)
    current = num[0]
    while left <= right and right < length:
        if current == n:
            count+=1
            current-=num[left]
            left+=1
        elif current > n:
            current-=num[left]
            left+=1
        elif current < n:
            right+=1
            if right < length:
                current+=num[right]

    print(count)
```

```python
import sys
from bisect import bisect_right

def arr_sums(arr):
    sums = [0]
    for weight in arr:
        sums+=[s + weight for s in sums]
    return sums

n, c = map(int, sys.stdin.readline().split())
    
num = sorted(list(map(int, sys.stdin.readline().split())))

group_a = num[:n//2]
group_b = num[n//2:]

sum_a = arr_sums(group_a)
sum_b = arr_sums(group_b)

sum_b.sort()

count = 0

for a in sum_a:
    if c-a<0:
        continue
    count+=bisect_right(sum_b, c-a)
    
print(count)
```
