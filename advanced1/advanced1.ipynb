```python
#3003
while 1:
    try:
        ref = [1, 1, 2, 2, 2, 8]
        arr = list(map(int, input().split()))
        final = [a-b for a, b in zip(ref, arr)]
        
        for i in range(6):
            print(final[i], end=" ")
            
    except Exception:
        break
```

```python
#10988
while 1:
    try:
        arr = list(input())
        l = len(arr)
        count = 0
        for i in range(l):
            if(arr[i]!=arr[l-1-i]):
                count+=1
        if(count == 0):
            print(1)
        else:
            print(0)
            
    except Exception:
        break
```

```python
#1157
while 1:
    try:
        word = input().upper()
        arr = list(set(word))
        cnt_list = []
        for x in arr:
            cnt = word.count(x)
            cnt_list.append(cnt)
            
        maximum = max(cnt_list)
        if(cnt_list.count(maximum) > 1):
            print("?")
        else:
            print(arr[cnt_list.index(maximum)])
            
    except Exception:
        break
```

```python
#2941
while 1:
    try:
        word = input()
        init_len = len(word)
        final_len = init_len
        cnt_list = ["c=", "c-", "d-", "lj", "nj", "s=", "z="]
        cnt = word.count("dz=")
        final_len = final_len - cnt
        for x in cnt_list:
            cnt = word.count(x)
            final_len = final_len - cnt
            
        print(final_len)
            
    except Exception:
        break
```
