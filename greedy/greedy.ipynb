```python
import sys

n, k = map(int, sys.stdin.readline().split())

coin = [int(sys.stdin.readline()) for _ in range(n)]

count = 0

for i in range(n-1, -1, -1):
    coef = 0
    if coin[i] <= k:
        coef = k // coin[i]
        k = k - coef*coin[i]
        count+=coef
    if k == 0:
        break
        
print(count)
```

```python
import sys

n = int(sys.stdin.readline())

time = [list(map(int, sys.stdin.readline().split())) for _ in range(n)]

sorted_time = sorted(time, key=lambda x: [x[1], x[0]])

end_time = 0
count = 0

for i in range(n):
    if end_time <= sorted_time[i][0]:
        count+=1
        end_time = sorted_time[i][1]
        
print(count)
```

```python
import sys

n = int(sys.stdin.readline())

time = list(map(int, sys.stdin.readline().split()))

sorted_time = sorted(time)

for i in range(1, n):
    sorted_time[i] += sorted_time[i-1]

print(sum(sorted_time))
```

```python
import sys

equation = sys.stdin.readline().rstrip().split("-")
equation = [list(map(int, word.split("+"))) for word in equation]
# print(equation)

numbers = []
for line in equation:
    numbers.append(sum(line))
    
# print(numbers)
length = len(numbers)
total = numbers[0]
for i in range(1, length):
    total-=numbers[i]
    
print(total)
```

```python
import sys

n = int(sys.stdin.readline())
road = list(map(int, sys.stdin.readline().split()))
city = list(map(int, sys.stdin.readline().split()))

min_price = city[0]
for i in range(1, n-1):
    if min_price < city[i]:
        city[i] = min_price
    else:
        min_price = city[i]
total = 0
for i in range(n-1):
    total+=city[i]*road[i]

print(total)
```
