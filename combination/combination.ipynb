```python
n, k = map(int, input().split())

denominator = 1
numerator = 1

for i in range(n-k+1, n+1):
    numerator = numerator*i

for i in range(1, k+1):
    denominator = denominator*i
    
print(numerator//denominator)
```

```python
T = int(input())

for _ in range(T):
    k, n = map(int, input().split())

    denominator = 1
    numerator = 1

    for i in range(n-k+1, n+1):
        numerator = numerator*i

    for i in range(1, k+1):
        denominator = denominator*i
        
    print(numerator//denominator)
```
