```python
while 1:
    try:
        arr_x=[0]*1001
        arr_y=[0]*1001
        for _ in range(3):
            x, y = map(int, input().split())
            arr_x[x]+=1
            arr_y[y]+=1
        print(arr_x.index(1), arr_y.index(1))
        
    except Exception:
        break
```

```python
while 1:
    try:
        arr_x=[]
        arr_y=[]
        n = int(input())
        for _ in range(n):
            x, y = map(int, input().split())
            arr_x.append(x)
            arr_y.append(y)
        w = max(arr_x)-min(arr_x)
        h = max(arr_y)-min(arr_y)
        print(w*h)
        
    except Exception:
        break
```

```python
while 1:
    try:
        a = int(input())
        b = int(input())
        c = int(input())
        if a+b+c!=180:
            print("Error")
        elif a == 60 and b==60 and c==60:
            print("Equilateral")
        elif a == b or b==c or a==c:
            print("Isosceles")
        else:
            print("Scalene")
        
    except Exception:
        break
```

```python
while 1:
    try:
        while 1:
            a, b, c = map(int, input().split())
            
            if a==0 and b == 0 and c==0:
                break

            arr = sorted([a, b, c])
            if(arr[2]>=arr[1]+arr[0]):
                print("Invalid")
                break
            
            unique_count=len({a, b, c})
            if unique_count==1:
                print("Equilateral")
            elif unique_count==2:
                print("Isosceles")
            else:
                print("Scalene")
        
    except Exception:
        break
```
