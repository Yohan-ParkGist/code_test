```python
while 1:
    try:
        a, b = map(int, input().split())
        print(a+b)
        print(a-b)
        print(a*b)
        print(int(a/b))
        print(a%b)
    except EOFError:
        break
```

```python
while 1:
    try:
        a = input()
        print(int(a)-543)
    except EOFError:
        break
```

```python
while 1:
    try:
        a, b, c = map(int, input().split())
        print((a+b)%c)
        print((a%c + b%c)%c)
        print((a*b)%c)
        print((a%c * b%c)%c)
    except EOFError:
        break
```

```python
while 1:
    try:
        a = int(input())
        b = int(input())
        print(a*(b%10))
        print(int(a*(b%100-b%10)/10))
        print(a*int(b/100))
        print(a*b)
    except Exception:
        break
```
