```python
while 1:
    try:
        a, b = map(int, input().split())
        if (a>b):
            print(">")
        elif(a<b):
            print("<")
        else:
            print("==")

    except Exception:
        break
```

```python
while 1:
    try:
        a = int(input())
        if (a>=90):
            print("A")
        elif(a>=80):
            print("B")
        elif(a>=70):
            print("C")
        elif(a>=60):
            print("D")
        else:
            print("F")

    except Exception:
        break
```

```python
while 1:
    try:
        a, b = map(int, input().split())
        c = int(input())
        h = c//60
        m = c%60
        a+=h
        b+=m
        
        if(b>=60):
            a+=1
            if(a>=24):
                print(a-24, b-60)
            else:
                print(a, b-60)
        else:
            if(a>=24):
                print(a-24, b)
            else:
                print(a, b)
        
        
    except Exception:
        break
```
