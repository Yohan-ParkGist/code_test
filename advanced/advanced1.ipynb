```python
#10988
while 1:
    try:
        arr = list(input())
        l = len(arr)
        count = 0
        for i in range(l):
            if(arr[i]!=arr[l-1-i]):
                count+=1
        if(count == 0):
            print(1)
        else:
            print(0)
            
    except Exception:
        break
```

```python
#1157
while 1:
    try:
        word = input().upper()
        arr = list(set(word))
        cnt_list = []
        for x in arr:
            cnt = word.count(x)
            cnt_list.append(cnt)
            
        maximum = max(cnt_list)
        if(cnt_list.count(maximum) > 1):
            print("?")
        else:
            print(arr[cnt_list.index(maximum)])
            
    except Exception:
        break
```

```python
#2941
while 1:
    try:
        word = input()
        init_len = len(word)
        final_len = init_len
        cnt_list = ["c=", "c-", "d-", "lj", "nj", "s=", "z="]
        cnt = word.count("dz=")
        final_len = final_len - cnt
        for x in cnt_list:
            cnt = word.count(x)
            final_len = final_len - cnt
            
        print(final_len)
            
    except Exception:
        break
```

```python
while 1:
    try:
        n = int(input())
        count = n
        for _ in range(n):
            word = input()
            for i in range(len(word)-1):
                if word[i]!=word[i+1]:
                    if word[i] in word[i+1:]:
                        count-=1
                        break
        print(count)
    except Exception:
        break
```

```python
while 1:
    try:
        score = {
            'A+': 4.5,
            'A0': 4.0,
            'B+': 3.5,
            'B0': 3.0,
            'C+': 2.5,
            'C0': 2.0,
            'D+': 1.5,
            'D0': 1.0,
            'F': 0.0
        }
        total_credit = 0.0
        total_score = 0.0
        for _ in range(20):
            name, credit, grade = input().split()
            credit = float(credit)
            if grade == 'P':
                continue
            else:
                total_score+=score[grade]*credit
                total_credit+=credit
            
        print(total_score/total_credit)

    except Exception:
        break
```
