```python
import sys

n = int(input())
name = []
count = 0
for i in range(n):
    message = sys.stdin.readline().rstrip()
    name.append(message)
    if message == "ENTER":
        count = count + len(set(name))-1
        name = []
    elif i==n-1:
        count = count + len(set(name))
        
print(count)
```

```python
import sys

n = int(input())

user_state = {"ChongChong":"dance"}
dancing = {"dance": {"ChongChong"}}


for i in range(n):
    one, two = sys.stdin.readline().split()
    if user_state.get(one) == "dance" or user_state.get(two) == "dance":
        user_state[one] = "dance"
        user_state[two] = "dance"
        dancing["dance"].add(one)
        dancing["dance"].add(two)
    
print(len(dancing["dance"]))
```

```python
import sys

n = int(sys.stdin.readline())

arr = []
count_list = [0] * 8001

for _ in range(n):
    num = int(sys.stdin.readline())
    arr.append(num)
    count_list[num+4000]+=1
    
max_freq = max(count_list)
modes = []
for i in range(8001):
    if count_list[i] == max_freq:
        modes.append(i-4000)

list_sorted = sorted(arr)

mean = round(sum(arr) / n)
median = list_sorted[n//2]
list_range = list_sorted[-1]-list_sorted[0]

print(mean)
print(median)
if len(modes) == 1:
    print(modes[0])
else:
    print(modes[1])
print(list_range)
```

```python
import sys

n, m = map(int, sys.stdin.readline().split())
word_freq = {}
for _ in range(n):
    word = sys.stdin.readline().rstrip()
    if len(word)>=m:
        word_freq[word] = word_freq.get(word, 0) + 1
        
sorted_words = sorted(word_freq.keys(), key=lambda x: (-word_freq[x], -len(x), x))

for word in sorted_words:
    print(word)
```
