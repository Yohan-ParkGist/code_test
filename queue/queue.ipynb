```python
import sys

class stack():
    def __init__(self):
        self.stack_arr = []
        
    def push(self, x):
        self.stack_arr.append(x)

    def pop(self):
        if self.is_empty():
            return -1
        else:
            return self.stack_arr.pop()
        
    def is_empty(self):
        return len(self.stack_arr) == 0


while 1:
    s = stack()
    is_valid = "True"
    command = sys.stdin.readline().rstrip()
    if command[0] == ".":
        break
    
    for x in command:
        if x == "(":
            s.push(x)
        elif x == ")":
            check = s.pop()
            if check == -1 or check=="[":
                is_valid = "False"
                break
        elif x == "[":
            s.push(x)
        elif x == "]":
            check = s.pop()
            if check == -1 or check=="(":
                is_valid = "False"
                break
        elif x == ".":
            break
    
    if is_valid == "False":
        print("no")
    else:
        if s.is_empty():
            print("yes")
        else:
            print("no")
```

```python
import sys
from collections import deque

class queue():
    def __init__(self):
        self.deq = deque()
    def push(self, x):
        self.deq.append(x)
    def pop(self):
        if self.empty():
            return -1
        else:
            return self.deq.popleft()
    def size(self):
        return len(self.deq)
    def empty(self):
        return len(self.deq) == 0
    def front(self):
        if self.empty():
            return -1
        else:
            return self.deq[0]
    def back(self):
        if self.empty():
            return -1
        else:
            return self.deq[-1]

n = int(sys.stdin.readline().rstrip())
q = queue()
for _ in range(n):
    command = sys.stdin.readline().split()
    if command[0] == 'push':
        q.push(int(command[1]))
    elif command[0] == 'pop':
        print(q.pop())
    elif command[0] == 'size':
        print(q.size())
    elif command[0] == 'empty':
        print(int(q.empty()))
    elif command[0] == 'front':
        print(q.front())
    elif command[0] == 'back':
        print(q.back())
```

```python
import sys
from collections import deque


n, m = map(int, sys.stdin.readline().split())
q = deque(range(1, n+1))
result = []

while 1:
    if len(q)==0:
        break
    else:
        q.rotate(-(m-1))
        result.append(str(q.popleft()))
        
print("<" + ", ".join(result) + ">")
```

```python
import sys
from collections import deque


n = int(sys.stdin.readline())
q = deque(enumerate(map(int, sys.stdin.readline().split()), start = 1))
result = []

while 1:
    if len(q)==0:
        break
    else:
        index, next = q.popleft()
        if next > 0:
            q.rotate(-(next-1))
        else:
            q.rotate(-next)
        result.append(index)
        
print(*result)
```
