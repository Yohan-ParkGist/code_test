```python
import sys

class max_heap():
    def __init__(self):
        self.heap_arr = [None]
    
    def insert(self, x):
        self.heap_arr.append(x)
        current_idx = len(self.heap_arr) - 1
        while current_idx > 1:
            parent = current_idx//2
            if self.heap_arr[current_idx] > self.heap_arr[parent]:
                self.heap_arr[current_idx], self.heap_arr[parent] = self.heap_arr[parent], self.heap_arr[current_idx]
                current_idx = parent
            else:
                break
        
    def pop(self):
        if len(self.heap_arr) <= 1:
            return 0
        to_return = self.heap_arr[1]
        if len(self.heap_arr) == 2:
            return self.heap_arr.pop()
        
        self.heap_arr[1] = self.heap_arr.pop()
        self.heapify(1)
        return to_return
        
    def heapify(self, i):
        size = len(self.heap_arr) - 1
        left = 2*i
        right = 2*i+1
        largest = i
        if left<=size and self.heap_arr[left]>self.heap_arr[largest]:
            largest = left
        if right<=size and self.heap_arr[right]>self.heap_arr[largest]:
            largest = right
        if largest!=i:
            self.heap_arr[i], self.heap_arr[largest] = self.heap_arr[largest], self.heap_arr[i]
            self.heapify(largest)

n = int(sys.stdin.readline())
heap = max_heap()

for _ in range(n):
    order = int(sys.stdin.readline())
    if order == 0:
        print(heap.pop())
    else:
        heap.insert(order)
```

```python
import sys
import heapq

def median(arr, arr_len):
    min_heap = [arr[0]]
    min_heap_len = 1
    max_heap = []
    max_heap_len = 0
    result = [min_heap[0]]
    for i in range(1, arr_len):
        if arr[i] >= min_heap[0]:
            heapq.heappush(min_heap, arr[i])
            min_heap_len+=1
        else:
            heapq.heappush(max_heap, -arr[i])
            max_heap_len+=1
            
        if i % 2 == 0:
            diff = min_heap_len - max_heap_len
            if diff == 1:
                pass
            elif diff == -1:
                heapq.heappush(min_heap, -heapq.heappop(max_heap))
                min_heap_len += 1
                max_heap_len -= 1
            elif diff == 3:
                heapq.heappush(max_heap, -heapq.heappop(min_heap))
                min_heap_len -= 1
                max_heap_len += 1
                
            result.append(min_heap[0])
            
    ans_len = len(result)
    print(ans_len)
    for i in range(ans_len):
        if (i+1)%10==0 or i == (ans_len-1):
            print(result[i])
        else:
            print(result[i], end=" ")
            
t = int(sys.stdin.readline())

for _ in range(t):
    n = int(sys.stdin.readline())
    if n%10 == 0:
        line = n%10
    else:
        line = n // 10 + 1
    num_list = []
    for _ in range(line):
        num_list = num_list + list(map(int, sys.stdin.readline().split()))
    
    median(num_list, n)
```
