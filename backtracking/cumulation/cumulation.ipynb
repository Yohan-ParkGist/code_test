```python
import sys

n, m = map(int, sys.stdin.readline().split())
num_list = [0]
num_list = num_list + list(map(int, sys.stdin.readline().split()))

for i in range(1, n+1):
    num_list[i] = num_list[i]+num_list[i-1]

for _ in range(m):
    i, j = map(int, sys.stdin.readline().split())
    print(num_list[j] - num_list[i-1])
```

```python
import sys

n, k = map(int, sys.stdin.readline().split())
num_list = list(map(int, sys.stdin.readline().split()))
cummulative_sum = [0]*(n-k+1)

for i in range(0, k):
    cummulative_sum[0] += num_list[i]

for i in range(1, n-k+1):
    cummulative_sum[i] = cummulative_sum[i-1] - num_list[i-1] + num_list[i+k-1]
    
print(max(cummulative_sum))
```

```python
import sys

word = sys.stdin.readline().rstrip()

word_len = len(word)
alphabet = [[0]*(word_len) for _ in range(26)]
for i in range(word_len):
    for j in range(26):
        alphabet[j][i] = alphabet[j][i-1]
    char_idx = ord(word[i])-97
    alphabet[char_idx][i] += 1

n = int(sys.stdin.readline())

for _ in range(n):
    a, i, j = sys.stdin.readline().split()
    char_idx = ord(a)-97
    if int(i) == 0:
        print(alphabet[char_idx][int(j)])
    else:
        print(alphabet[char_idx][int(j)]-alphabet[char_idx][int(i)-1])
```

```python
import sys

n, m = map(int, sys.stdin.readline().split())
num_list = list(map(int, sys.stdin.readline().split()))

for i in range(1, n):
    num_list[i] = num_list[i] + num_list[i-1]

remainder_count = [0]*m
for i in range(n):
    num_list[i] = num_list[i]%m
    remainder_count[num_list[i]]+=1

total = 0
for i in range(m):
    total += remainder_count[i]*(remainder_count[i]-1)//2
    
print(total+remainder_count[0])
```

```python
import sys

n, m = map(int, sys.stdin.readline().split())
num_list = [[0] * (n + 1)] + [[0] + list(map(int, input().split())) for _ in range(n)]

for i in range(1, n+1):
    for j in range(1, n+1):
        num_list[i][j] = num_list[i][j] + num_list[i-1][j] + num_list[i][j-1] - num_list[i-1][j-1]

for i in range(m):
    x1, y1, x2, y2 = map(int, sys.stdin.readline().split())
    print(num_list[x2][y2]+num_list[x1-1][y1-1]-num_list[x1-1][y2]-num_list[x2][y1-1])
```

```python
import sys

n, m, k = map(int, sys.stdin.readline().split())
board = [[0] * (m + 1)] + [[0] + list(sys.stdin.readline().rstrip()) for _ in range(n)]

for i in range(1, n+1):
    for j in range(1, m+1):
        if (i+j)%2==0:
            if board[i][j] == "W":
                board[i][j] = 0
            else:
                board[i][j] = 1
        else:
            if board[i][j] == "B":
                board[i][j] = 0
            else:
                board[i][j] = 1

for i in range(1, n+1):
    for j in range(1, m+1):
        board[i][j] = board[i][j] + board[i-1][j] + board[i][j-1] - board[i-1][j-1]

current_max = 0
current_min = k*k

for i in range(k, n+1):
    for j in range(k, m+1):
        current_sum = board[i][j]+board[i-k][j-k]-board[i-k][j]-board[i][j-k]
        if current_sum < current_min:
            current_min= current_sum
        if current_sum > current_max:
            current_max = current_sum

print(min(current_min, k*k-current_max))
```
