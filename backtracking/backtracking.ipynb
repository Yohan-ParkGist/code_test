```python
import sys

n, m = map(int, sys.stdin.readline().split())

visiting = [False]*(n+1)

def generate_combinations_1(n, k):
    results = []
    combination = []
    def backtrack():
        if len(combination) == k:
            results.append(combination[:])
            return 

        for i in range(1, n+1):
            if not visiting[i]:
                visiting[i] = True
                combination.append(i)
                backtrack()
                combination.pop()
                visiting[i] = False

    backtrack()
    return results

final_result = generate_combinations_1(n, m)

for line in final_result:
    print(*line)
```

```python
import sys

board = []

for _ in range(9):
    board.append(list(map(int, sys.stdin.readline().split())))

zero_index = []

for i in range(9):
    for j in range(9):
        if board[i][j]==0:
            zero_index.append((i, j))

def check_col_row(value, r, c):
    if value in board[r]:
        return False
    if value in [board[i][c] for i in range(9)]:
        return False
    return True
    
def check_square(value, r, c):
    start_row = (r//3)*3
    start_col = (c//3)*3
    for i in range(3):
        for j in range(3):
            if value == board[start_row+i][start_col+j]:
                return False
    return True
    
def backtrack(index):
    if index == len(zero_index):
        for line in board:
            print(*line)
        sys.exit(0)
    
    r, c = zero_index[index]
    
    for value in range(1, 10):
        if check_col_row(value, r, c) and check_square(value, r, c):
            board[r][c]=value
            backtrack(index+1)
            board[r][c]=0


backtrack(0)
```

```python
import sys

n = int(sys.stdin.readline())
num_list = list(map(int, sys.stdin.readline().split()))
plus, minus, multi, div = map(int, sys.stdin.readline().split())

operator = []
results = []
total = 0
def dfs(total, index, plus, minus, multi, div):
    if index == n:
        results.append(total)
        return
    
    if plus>0:
        dfs(total+ num_list[index], index+1, plus-1, minus, multi, div)
    if minus>0:
        dfs(total-num_list[index], index+1, plus, minus-1, multi, div)
    if multi>0:
        dfs(total*num_list[index], index+1, plus, minus, multi-1, div)
    if div>0:
        dfs(int(total / num_list[index]), index+1, plus, minus, multi, div-1)
        
dfs(num_list[0], 1, plus, minus, multi, div)
print(max(results))
print(min(results))
```

```python
import sys

n = int(sys.stdin.readline())
people_matrix = [list(map(int, sys.stdin.readline().split())) for _ in range(n)]

result = []
combination = []
def dfs(start):
    if len(combination) == n//2:
        result.append(combination[:])
        return
    
    for i in range(start, n):
        combination.append(i)
        dfs(i+1)
        combination.pop()
        
dfs(0)
# print(result)
score_list = []
for line in result:
    team_score = 0
    for i in range(n//2):
        for j in range(i+1, n//2):
            team_score+=people_matrix[line[i]][line[j]]
            team_score+=people_matrix[line[j]][line[i]]
    score_list.append(team_score)

minimum_difference = abs(score_list[0] - score_list[-1])
length = len(score_list)
for i in range(length//2):
    diff = abs(score_list[i]-score_list[length-i-1])
    if minimum_difference > diff:
        minimum_difference = diff

# print(score_list)
print(minimum_difference)
```
