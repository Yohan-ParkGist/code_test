```python
import sys

n, m = map(int, sys.stdin.readline().split())

visiting = [False]*(n+1)

def generate_combinations_1(n, k):
    results = []
    combination = []
    def backtrack():
        if len(combination) == k:
            results.append(combination[:])
            return 

        for i in range(1, n+1):
            if not visiting[i]:
                visiting[i] = True
                combination.append(i)
                backtrack()
                combination.pop()
                visiting[i] = False

    backtrack()
    return results

final_result = generate_combinations_1(n, m)

for line in final_result:
    print(*line)
```
