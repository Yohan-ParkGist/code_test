```python
while 1:
    try:
        A = 1
        B = 1
        while(A!=0 and B!=0):
            A, B = map(int, input().split())

            if B%A==0:
                print("factor")
            elif(A%B==0):
                print("multiple")
            else:
                print("neither")
            
    except Exception:
        break
```

```python
while 1:
    try:

        N, K = map(int, input().split())
        cnt = 0
        for i in range(1, N+1):
            if(N%i==0):
                cnt+=1
            if(cnt==K):
                print(i)
                break
        if(K>cnt):
            print(0)
            
    except Exception:
        break
```

```python
while 1:
    try:
        while 1:
            divisor = []
            sum = 0
            n = int(input())
            if n==-1:
                break
            for i in range(2, (n//2)+2):
                if(n%i==0):
                    sum+=i
                    divisor.append(i)
            if(sum+1==n):
                print(f"{n} = 1", end="")
                for num in divisor:
                    print(f" + {num}", end="")
            else:
                print(f"{n} is NOT perfect.", end="")
            print()
    except Exception:
        break
```

```python
N = int(input())

i = 2
# i * i <= N 은 i <= sqrt(N)과 같은 의미입니다.
# 즉, i가 루트 N보다 커지면 반복을 종료합니다.
while i * i <= N:
    while N % i == 0:  # 해당 소수로 나누어 떨어질 때까지 계속 나눔
        print(i)
        N //= i
    i += 1

# 반복문이 끝났는데 N이 1이 아니라면?
# 남은 N은 그 자체로 소수라는 뜻입니다.
if N > 1:
    print(N)
```
