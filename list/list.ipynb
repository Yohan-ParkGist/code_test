```python
while 1:
    try:
        n = int(input())
        list = list(map(int, input().split()))
        v = int(input())
        count=0
        
        for i in range(n):
            if(list[i]==v):
                count+=1
        
        print(count)
            
    except Exception:
        break
```

```python
while 1:
    try:
        n, x = map(int, input().split())
        l = list(map(int, input().split()))
        
        for i in range(n):
            if(l[i] < x):
                print(l[i], end=" ")
            
    except Exception:
        break
```

```python
while 1:
    try:
        n = int(input())
        l = list(map(int, input().split()))
        min, max = l[0], l[0]
        for num in l:
            if (min > num):
                min=num
            if (max < num):
                max=num
        print(min, max)
            
    except Exception:
        break
```

```python
while 1:
    try:
        l = [int(input()) for _ in range(9)]
        max = l[0]
        loc = 0
        for num in l:
            if(max < num):
                max = num
        for i in range(9):
            if (max == l[i]):
                loc = i+1
                
        print(max)
        print(loc)
             
    except Exception:
        break
```

```python
while 1:
    try:
        n, m = map(int, input().split())
        arr = [0]*n
        for _ in range(m):
            i, j, k = map(int, input().split())
            for o in range(i-1, j):
                arr[o] = k
        for i in range(n):
            print(arr[i], end=" ")
    except Exception:
        break
```

```python
while 1:
    try:
        n, m = map(int, input().split())
        arr = list(range(n+1))
        for _ in range(m):
            i, j = map(int, input().split())
            temp = arr[i]
            arr[i] = arr[j]
            arr[j]= temp
        for i in range(n):
            print(arr[i+1], end=" ")
    except Exception:
        break
```

```python
while 1:
    try:
        arr = [0]*31
        for _ in range(28):
            i = int(input())
            arr[i] = 1
            
        for i in range(30):
            if (arr[i+1]==0):
                print(i+1)
                
    except Exception:
        break
```

```python
while 1:
    try:
        n, m = map(int, input().split())
        arr = list(range(1, n+1))
        for _ in range(m):
            i, j = map(int, input().split())
            arr[i-1:j]=arr[i-1:j][::-1]
            
        for i in range(n):
            print(arr[i], end=" ")
            
    except Exception:
        break
```

```python
while 1:
    try:
        n = int(input())
        arr = list(map(int, input().split()))
        max_val = max(arr)
        mean = sum(arr)/n
        
        print(mean*100/max_val)
    except Exception:
        break
```
