```python
# insertion sort

while 1:
    try:
        n = int(input())
        arr = []
        for _ in range(n):
            arr.append(int(input()))
        for i in range(1, n):
            for j in range(i, 0, -1):
                if arr[j]<arr[j-1]:
                    arr[j], arr[j-1] = arr[j-1], arr[j]
                else:
                    break
        for num in arr:
            print(num)

    except Exception:
        break
```

```python
# heap sort

def heapify(arr, n, i):
    left = 2*i+1
    right = 2*i+2
    largest = i
    if left<n and arr[left]>arr[largest]:
        largest = left
    if right<n and arr[right]>arr[largest]:
        largest = right
    if largest!=i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)
    
    
def heapsort(arr):
    n = len(arr)
    for i in range(n//2-1, -1, -1):
        heapify(arr, n, i)
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)
        
    
while 1:
    try:
        n = int(input())
        arr = []
        for _ in range(n):
            arr.append(int(input()))

        heapsort(arr)
        for num in arr:
            print(num)


    except Exception:
        break
```

```python
# Counting sort

def counting_sort(arr):
    counting_arr = [0]*10001
    for num in arr:
        counting_arr[num]+=1
        
    return counting_arr
    
while 1:
    try:
        n = int(input())
        arr = []
        for _ in range(n):
            arr.append(int(input()))

        sort = counting_sort(arr)
        for i in range(len(sort)):
            while sort[i]!=0:
                print(i)
                sort[i]-=1


    except Exception:
        break
```

```python
def heapify(arr, n, i):
    left = 2*i+1
    right = 2*i+2
    largest = i
    if left<n and arr[left]>arr[largest]:
        largest = left
    if right<n and arr[right]>arr[largest]:
        largest = right
    if largest!=i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)
    
    
def heapsort(arr):
    n = len(arr)
    for i in range(n//2-1, -1, -1):
        heapify(arr, n, i)
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)
        
    
while 1:
    try:
        number_string = input()
        digits = [int(num) for num in number_string]

        heapsort(digits)
        for i in range(len(digits)-1, -1, -1):
            print(digits[i], end="")

    except Exception:
        break
```

```python
import sys

n = int(input())
words = []
for _ in range(n):
    words.append(sys.stdin.readline().rstrip())
words = list(set(words))
sorted_list = sorted(words, key=lambda x: (len(x), x))
for word in sorted_list:
    print(word)
```
