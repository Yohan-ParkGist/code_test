```python
import sys

n = int(sys.stdin.readline())
card = set(map(int, sys.stdin.readline().split()))
m = int(sys.stdin.readline())
ans = list(map(int, sys.stdin.readline().split()))

for num in ans:
    if num in card:
        print(1, end=" ")
    else:
        print(0, end=" ")
```

```python
import sys

n, m = map(int, sys.stdin.readline().split())
s = {sys.stdin.readline().strip() for _ in range(n)}
check = [sys.stdin.readline().strip() for _ in range(m)]
count = 0

for words in check:
    if words in s:
        count+=1
        
print(count)
```

```python
import sys

n = int(sys.stdin.readline())
people = {}
for _ in range(n):
    name, status = sys.stdin.readline().split()
    people[name] = status

result = sorted([name for name, status in people.items() if status == "enter"], reverse=True)

for name in result:
    print(name)
```

```python
import sys

n, m = map(int, sys.stdin.readline().split())
name_to_id = {}
id_to_name = {}

for i in range(1, n+1):
    name = sys.stdin.readline().rstrip()
    name_to_id[name] = i
    id_to_name[str(i)] = name
    
check = [sys.stdin.readline().strip() for _ in range(m)]

for query in check:
    if query.isdigit():
        print(id_to_name[query])
    else:
        print(name_to_id[query])
```

```python
import sys

n = int(sys.stdin.readline())
cards = list(map(int, sys.stdin.readline().split()))
m = int(sys.stdin.readline())
check = list(map(int, sys.stdin.readline().split()))

card_dict = {}

for num in cards:
    card_dict[num] = card_dict.get(num, 0) + 1

for num in check:
    print(card_dict.get(num, 0), end=" ")
    
```

```python
import sys

n, m = map(int, sys.stdin.readline().split())
a = {sys.stdin.readline().strip() for _ in range(n)}
b = {sys.stdin.readline().strip() for _ in range(m)}
c = list(a&b)
        
print(len(c))

for name in sorted(c):
    print(name)
```

```python
import sys

word = sys.stdin.readline().rstrip()
substring=set()

for i in range(1, len(word)+1):
    for j in range(len(word)-i+1):
        substring.add(word[j:j+i])
        
print(len(substring))
```
