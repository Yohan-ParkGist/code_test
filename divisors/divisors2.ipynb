```python
import sys

def LCM(a, b):
    a, b = max(a, b), min(a, b)
    r = 1
    while 1:
        r = a%b
        if r==0:
            return b
        else:
            a = b
            b = r


a, b = map(int, sys.stdin.readline().split())
c, d = map(int, sys.stdin.readline().split())

denominator = b*d//LCM(b, d)
numerator = a*denominator//b + c*denominator//d

devisor = LCM(numerator, denominator)

print(numerator//devisor, denominator//devisor)
```

```python
import sys

def GCD(a, b):
    a, b = max(a, b), min(a, b)
    r = 1
    while 1:
        r = a%b
        if r==0:
            return b
        else:
            a = b
            b = r


n = int(sys.stdin.readline().rstrip())

arr = [int(sys.stdin.readline().rstrip()) for _ in range(n)]
junction = [arr[i+1]-arr[i] for i in range(n-1)]

MGCD = GCD(junction[0], junction[1])
for i in range(2, n-1):
    MGCD = GCD(junction[i], MGCD)

total_gap = (arr[-1]-arr[0])//MGCD

print(total_gap-n+1)
```

```python
import sys

def is_prime(n):
    if n <= 1:
        return False
    if n<=3:
        return True
    if n%2==0 or n%3==0:
        return False
    i=5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
        
    return True

n = int(sys.stdin.readline().rstrip())

for _ in range(n):
    test = int(sys.stdin.readline().rstrip())
    j = test
    while 1:
        if is_prime(j):
            print(j)
            break
        else:
            j+=1
```

```python
import sys

sieve = bytearray([1])*1000001
sieve[0] = sieve[1] = 0

for i in range(2, int(1000000**0.5)+1):
    if sieve[i]:
        for j in range(i*i, 1000001, i):
            sieve[j]=0

primes = [i for i in range(1000001) if sieve[i]]

n = int(sys.stdin.readline().rstrip())

for _ in range(n):
    test = int(sys.stdin.readline().rstrip())
    count = 0
    for p in primes:
        if p > test//2:
            break
        if sieve[test-p]:
            count+=1
    
    print(count)
```

```python
import sys

n = int(sys.stdin.readline().rstrip())

# 약수의 개수가 홀수인 경우에 창문이 열린다.
# = 제곱수 1, 4, 9, 16 ...

i = 1
while 1:
    if i*i > n:
        print(i-1)
        break
    elif i*i == n:
        print(i)
        break
    else:
        i+=1
```
